```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv("../data/wiki_corps/DAX_companies.csv")

# Display the first few rows of the DataFrame
df.head()
```

```python
from matplotlib import pyplot as plt
plt.bar(df.groupby("Sector").count().index, df.groupby("Sector").count()["Company"])
plt.xticks(rotation=90)
plt.ylabel("Number of Companies")
plt.title("Companies per Sector in DAX")
plt.show()
```
